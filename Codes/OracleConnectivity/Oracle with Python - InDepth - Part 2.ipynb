```python
import cx_Oracle
create_table = "create table hotel_dimension (total_bill number(10,2), tip number(10,2), sex varchar2(10), smoker varchar2(10), day varchar2(10), time varchar2(10), dsize number)"
con = cx_Oracle.connect('hr/atharv2210@localhost:1521/ORCL')
cur = con.cursor()
cur.execute('Drop table hotel_dimension')
cur.execute(create_table)
rows = [tuple(x) for x in df.values]
cur.executemany("insert into hotel_dimension (total_bill, tip, sex,smoker,day,time,dsize) values (:1,:2,:3,:4,:5,:6,:7)", rows)
con.commit()
cur.close()
con.close()
```

```python
import cx_Oracle
#create_table = "alter table hotel_dimension add  discount int"
con = cx_Oracle.connect('hr/atharv2210@localhost:1521/ORCL')
cur = con.cursor()
discount = int(input('Enter the discount'))
total_bill_range = float(input('Enter the total_bill range'))
query = "update hotel_dimension set discount = %d where total_bill > %f"
cur.execute(query %(discount, total_bill_range))
con.commit()
print('Records updated successfully')
cur.close()
con.close()
```
